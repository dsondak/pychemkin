```python
#Import SQLlite 
import sqlite3

#Connect to DB
db = sqlite3.connect('NASA7_coeffs.sqlite')
cursor = db.cursor()
```

```python
#Imports 
import numpy as np
import pandas as pd

#table visualization function 
def viz_tables(cols, query):
    q = cursor.execute(query).fetchall()
    framelist = []
    for i, col_name in enumerate(cols):
        framelist.append((col_name, [col[i] for col in q]))
    return pd.DataFrame.from_items(framelist)
```

```python
candidate_cols = [col[1] for col in cursor.execute("PRAGMA table_info(LOW)")]
query = '''SELECT * FROM LOW'''
viz_tables(candidate_cols, query)
```

```python
candidate_cols = [col[1] for col in cursor.execute("PRAGMA table_info(HIGH)")]
query = '''SELECT * FROM HIGH'''
viz_tables(candidate_cols, query)
```

```python
#Function to Get All Species
'''
This function will accept one parameter the temp range in the form of 'low' or 'high' as represented in the db
This function will return a list of species.
'''

def get_all_species(temp_range):
    #get the species based on temp range
    if temp_range == 'low':
        query = '''SELECT SPECIES_NAME FROM LOW'''
    elif temp_range == 'high':
        query = '''SELECT SPECIES_NAME FROM HIGH'''
    else:
        raise ValueError("Must input 'low' or 'high' for temp_range")
    
    species_list = cursor.execute(query).fetchall()
    species_list = [s[0].strip(',') for s in species_list]
    
    return species_list
```

```python
#Function to Get Coeffs
'''This function will accept two parameter the species name (as shown in the species name column of db tables) and
    the temp range in the form of 'low' or 'high' as represented in the db
'''
def get_coeffs(species_name, temp_range):
    #get coeffs based on temp_range
    if temp_range == 'low':
        #get list of species
        species_list = get_all_species('low')
        
        #check that specie_name inputed it correct type of species
        if species_name not in species_list:
            raise ValueError("Input must be a valid species_name. The species_name input doesn't exist!")
            
        #create query
        query = '''SELECT COEFF_1,COEFF_2,COEFF_3,COEFF_4,COEFF_5,COEFF_6,COEFF_7 FROM LOW WHERE SPECIES_NAME = ? '''
        
    elif temp_range == 'high':
        species_list = get_all_species('high')
        if species_name not in species_list:
            raise ValueError("Input must be a valid species_name. The species_name input doesn't exist!")
        
        #create query
        query = '''SELECT COEFF_1,COEFF_2,COEFF_3,COEFF_4,COEFF_5,COEFF_6,COEFF_7 FROM HIGH WHERE SPECIES_NAME = ? '''  
    else:
         raise ValueError("Must input 'low' or 'high' for temp_range")
    
    coeffs = cursor.execute(query,(species_name,)).fetchall()
    return coeffs
```

```python
#test get_coeffs function
print(get_coeffs('Ag','high'))
```
